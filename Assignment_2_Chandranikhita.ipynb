```python
##6  Outlier detection and replacement
sns.boxplot(df.Age)
```

```python
##IQR Method
q3=df.Age.quantile(0.75)
q1=df.Age.quantile(0.25)
IQR=q3-q1
upper_limit=q3+1.5*IQR
lower_limit=q1-1.5*IQR
upper_limit
```

```python
df['Age']= np.where(df['Age']>upper_limit,37,df['Age'])
df[10:20]    ##16th row of Age column is getting replaced from 58 to 37
```

```python
##Label encoding---binary categories
##one-hot encoding ---for multiple categories
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Gender=le.fit_transform(df.Gender)
df.HasCrCard=le.fit_transform(df.HasCrCard)
df.IsActiveMember=le.fit_transform(df.IsActiveMember)
df.Exited=le.fit_transform(df.Exited)
df.head()
```

```python
##One-hot Encoding
import pandas as pd
df_main=pd.get_dummies(df,columns=['Geography','NumOfProducts'])
df_main.head()
```

```python
df_main.shape
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df_main.corr(),annot=True)
```

```python
df_main.describe()
```

```python
##8 x and y split
```

```python
x = df_main[['Age','Tenure']].copy()
x
```

```python
y=df_main.Exited
y
```

```python
y.shape
```

```python
##9 scale the independent variables
```

```python
from sklearn.preprocessing import scale
x_scaled=pd.DataFrame(scale(x),columns=x.columns)
x_scaled
```

```python
##10 Train Test split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=0)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
##Regression ML
from sklearn.linear_model import LinearRegression
le= LinearRegression()
le.fit(x_train,y_train)
y_predict=le.predict(x_test)
y_predict
```

```python
pd.DataFrame({'Actual':y_test,'Predicted':y_predict})
```
